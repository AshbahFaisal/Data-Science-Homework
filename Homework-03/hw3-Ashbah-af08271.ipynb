```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
df = pd.read_csv('C:\\Users\\cw\\Downloads\\student-scores.csv')
df
```

### question 1 If math_score of males is greater than females
H0:math score of males and females are the same <br>
Ha:math score of males is greater than females

```python
male_scores = df[df['gender'] == 'male']['math_score']
female_scores = df[df['gender'] == 'female']['math_score']
#stats.ttest_ind(male, female)
stats.ttest_ind(male_scores, female_scores, alternative="greater")
```

### Conclusion
Since the pvalue<0.05 hence reject the null hypothesis and accept the alternative i.e math score of males is greater than females

### Question 2 If math_score of Non Doctors is greater than Doctors 
H0:math score of non doctors is the same as doctors <br>
Ha:math score of non doctors is greater than doctors

```python
non_doctor_scores = df[df['career_aspiration'] != 'Doctor']['math_score']
doctor_scores = df[df['career_aspiration'] == 'Doctor']['math_score']
stats.ttest_ind(non_doctor_scores, doctor_scores, alternative="greater")
```

### Conclusion
Since the pvalue>0.05 hence we accept the null hypothesis and reject the alternative hypothesis i.e the math score of the non doctors is equal to the math scores of the doctors.

### Question 4 
### If math_score of no part_time_job male students are greater than part_time_job male students<br>

### Question 9 Is there any Association between gender and career_aspiration
H0:gender and career_aspiration are independent <br>
Ha:gender and career_aspiration are dependent

```python
chisqt= pd.crosstab(df['gender'], df['career_aspiration'], margins=True)
stats.chi2_contingency(chisqt)
chi2, p, dof, ex = stats.chi2_contingency(chisqt)
p
```

### Conclusion
Since the pvalue < 0.05  hence the null hypothesis is rejected and the alternative hypothesis is accepted i.e gender and career_aspirations are dependent
