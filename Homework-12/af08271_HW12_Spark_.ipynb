# Creating RDD (Resilient Distributed Dataset) - fundamental data structure of Spark

```python
sc = SparkContext.getOrCreate()
```

```python
products = sc.textFile("SalesData.txt")
```

```python
products.first()
```

```python
print("The data has {} columns".format(len(products.first().split("\t"))))
header = products.first().split("\t")
header
```

```python
head = products.first()
content = products.filter(lambda line: line != head)
```

```python
content.first()
```

```python
content.map(lambda line: len(line.split("\t"))).distinct().collect()[0]
```

# Example of RDD using Scala

```python
rdd1 = (content.filter(lambda line: line.split("\t"))
 .map(lambda line: (line.split("\t")))
)
rdd1.take(1)
```

# Converting RDD into Spark Dataframe

```python
df = rdd1.toDF(header)
df.show()
```

#### 5. Visualize Average List Price per Color and Average List Price per Size. Discuss/Explain your results.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pyspark.sql import functions as F
%matplotlib inline
```

```python
df1 = df.select(df.Name,df.ProductNumber, df.Color,df.Size,df.ListPrice.cast("float"))
df2= df1.groupBy(df.Color).agg({"ListPrice":"average"}).collect()
a=[]
b =[]
for i in df2:
  if i["Color"]!= "NULL":
    a.append(i["Color"])
    b.append(i["avg(ListPrice)"])
plt.figure(figsize=(10, 6))
plt.bar(a, b)
plt.title('Average List Price by Color')
plt.xlabel('Color')
plt.ylabel('Average List Price')
plt.show()
```

Based on the chart, the highest average list price is for Red, while the lowest is for White, with Red falling in between. The median price appears to be around the price point for Black, indicating that most products tend to be priced in this range, with fewer products priced as high as those in Red or as low as those in White. This suggests that Red might be a premium color, while White is a more standard or budget-friendly option. But its quite weird how just due to difference in color a products prices is so high or so low.

```python
df2= df1.groupBy(df.Size).agg({"ListPrice":"average"}).collect()
a=[]
b =[]
for i in df2:
  if i["Size"]!= "NULL":
    a.append(i["Size"])
    b.append(i["avg(ListPrice)"])

plt.bar(a, b)
plt.title('Average List Price by Size')
plt.xlabel('Size')
plt.ylabel('Average List Price')
plt.show()
```

The graph reveals that the average list price of size 56 is the highest, which is unexpected as larger sizes typically cost more. This suggests that the company leveraged the popularity and demand for size 56 to increase its price, while the price for size 70 is lower due to its reduced market demand. Furthermore, specific sizes tend to be priced higher than generic ones, reflecting the additional cost for customization, with customers paying a premium for these tailored options.







#### 6. Visualize Total Count per Size. Discuss/Explain your results.

```python
df2= df1.groupBy(df.Size).agg({"ListPrice":"count"}).collect()
a=[]
b =[]
for i in df2:
  if i["Size"]!= "NULL":
    a.append(i["Size"])
    b.append(i["count(ListPrice)"])

plt.bar(a, b)
plt.title('Count per Size')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

The graph shows the distribution of sizes, with size 44 having the highest count, suggesting that it is a more general or standard size, likely produced in larger quantities to meet broader demand. On the other hand, size 56, which had the highest average list price in the previous graph, has a relatively smaller count. This indicates that the company may be creating a monopoly by limiting the availability of this size, taking advantage of its scarcity to increase its price.

# Using Machine Learning Library (For Practice)

```python
dfall = df.select(df.Weight.cast("float"), df.Size.cast("float"), df.ListPrice.cast("float"))
dfall.show(5, truncate=False)
```

```python
dfall = dfall.na.drop("any")
```

```python
feature_columns = dfall.columns[:-1] # here we omit the final column for prediction
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=feature_columns,outputCol="features")
```

```python
data_2 = assembler.transform(dfall)
```

```python
data_2.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
algo = LinearRegression(featuresCol="features", labelCol="ListPrice")
```

```python
model = algo.fit(data_2)
```

```python
evaluation_summary = model.evaluate(data_2)
```

```python
evaluation_summary.rootMeanSquaredError
```
