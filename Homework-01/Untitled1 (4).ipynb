```python
#part_2

import pandas as pd

# Load the DataFrame
df = pd.read_csv('C:/Users/cw/Downloads/formatted580SurveyCleanup2.csv')

# Print columns with NaN values
cols_with_nan = df.columns[df.isnull().any()]
print(cols_with_nan)

# Using mode for columns where highest repeated value is applicable
df['Expected Graduation date from Mason MS program?'] = df['Expected Graduation date from Mason MS program?'].fillna(df['Expected Graduation date from Mason MS program?'].mode()[0])
df['What is your approximate current level of programming skill in Python?'] = df['What is your approximate current level of programming skill in Python?'].fillna(df['What is your approximate current level of programming skill in Python?'].mode()[0])
df['What is your approximate current level of programming skill in R?'] = df['What is your approximate current level of programming skill in R?'].fillna(df['What is your approximate current level of programming skill in R?'].mode()[0])
df['What is your approximate current level of programming skill in SQL?'] = df['What is your approximate current level of programming skill in SQL?'].fillna(df['What is your approximate current level of programming skill in SQL?'].mode()[0])

# using unknown where mean and mode make no sense
df['Region'] = df['Region'].fillna('Unknown')
df['City / Town / District'] = df['City / Town / District'].fillna('Unknown')
df['Country of Citizenship'] = df['Country of Citizenship'].fillna('Unknown')
df['Undergraduate Degree'] = df['Undergraduate Degree'].fillna('Unknown')
df['Undergraduate Degree'] = df['Undergraduate Degree'].fillna('Unknown')
df['What is your goal for learning about data analytics?'] = df['What is your goal for learning about data analytics?'].fillna('Unknown')

#where average is applicable
df['Height (Inches)'] = pd.to_numeric(df['Height (Inches)'], errors='coerce') # Convert to numeric, coerce errors to NaN
df['Height (Inches)'] = df['Height (Inches)'].fillna(df['Height (Inches)'].mean())

#clean age
current_year = datetime.now().year

def process_age(value):
    value = value.strip()
    
    if len(value) == 3: 
        return int(''.join(value.split()))     
    if len(value) == 4 and value.isdigit():
        return current_year - int(value)
    return pd.to_numeric(value, errors='coerce')
df['Age (years)'] = df['Age (years)'].apply(process_age).astype('Int64')
#since the options given are not the answer
df['Type of Laptop Used?'] = df['Type of Laptop Used?'].fillna('Other')
df['Are you employed full-time while attending Mason?'] = df['Are you employed full-time while attending Mason?'].fillna('Other')

df=df.drop (columns = ['Access Code','Email Address'])


df
df.to_csv("C:/Users/cw/Downloads/cleaned580Cleanup.csv")
```
