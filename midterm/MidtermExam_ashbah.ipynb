Q21. Test the Hypothesis and explain your findings: <br> H0: the score of **type** = TV Show and **type** = Movie is same (equal or similar). <br>
Ha: the score of **type** = TV Show is greater than **type** = Movie <br>
**Briefly explain your conclusion**

```python
tv_show_scores = final_combined_df[final_combined_df['type'] == 'TV Show']['score']
movie_scores =final_combined_df[final_combined_df['type'] == 'Movie']['score']
stats.ttest_ind(tv_show_scores, movie_scores, alternative='greater')

```

### Conclusion
Since the p-value is greater than 0.05 accept the null hypothesis i.e TV show and movie types score is equal or similar.

Q22. Test the Hypothesis and explain your findings: <br> H0: **type** and **country** are independent and are not related to each other <br>
Ha: **type** and **country** are dependent and related to each other <br>
**Briefly explain your conclusion**

```python
chisqt= pd.crosstab(final_combined_df['type'], final_combined_df['country'], margins=True)
stats.chi2_contingency(chisqt)
chi2, p, dof, ex = stats.chi2_contingency(chisqt)
p
```

### Conclusion
since the pvalue<0.05 hence the alternative hypothesis is accepted i.e type and country are dependent on each other
